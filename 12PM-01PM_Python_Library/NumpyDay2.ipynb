```python
# InterQuantileRange
Q1 = df["Unit Price"].quantile(0.25)
Q3 = df["Unit Price"].quantile(0.75)
IQR = Q3-Q1
min_range = Q1-(1.5*IQR)
max_range = Q3+(1.5*IQR)
min_range,max_range
```

```python
df = df[df["Unit Price"]<max_range]
```
